```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv("./train.csv")

titanic_df.head(5)
```

## 1. 전체 생존율 (원형 그래프)

```python
s = titanic_df[titanic_df.Survived == 1].Survived.count()
d = titanic_df[titanic_df.Survived == 0].Survived.count()

plt.pie([s,d], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
plt.show()
```

* 생존율 : 38.38%
* 사망율 : 61.62%

대략 10명 중 6명 사망.

## 2. 각 티켓 클래스별 생존율 (원형 그래프)

```python
p1 = titanic_df[titanic_df.Pclass==1].Survived.mean()
p2 = titanic_df[titanic_df.Pclass==2].Survived.mean()
p3 = titanic_df[titanic_df.Pclass==3].Survived.mean()

fig = plt.figure()
fig.suptitle("Survival rate per ticket class", x=1.2, y=2.2, fontsize=20)
ax1 = fig.add_axes([0, 1, 1, 1], aspect=1)
ax1.title.set_text("Class 1")
ax2 = fig.add_axes([.8, 1, 1, 1], aspect=1)
ax2.title.set_text("Class 2")
ax3 = fig.add_axes([1.6, 1, 1, 1], aspect=1)
ax3.title.set_text("Class 3")
ax1.pie([p1, 1-p1], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
ax2.pie([p2, 1-p2], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
ax3.pie([p3, 1-p3], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
plt.show()
```

- Class 1: 62.96%
- Class 2: 47.28%
- Class 3: 24.24%

티켓 등급이 높을수록 생존율이 높았다.

## 3. 성별에 따른 생존율 (원형 그래프)

```python
m = titanic_df[titanic_df.Sex=='male'].Survived.mean()
f = titanic_df[titanic_df.Sex=='female'].Survived.mean()

fig = plt.figure()
fig.suptitle("Survival rate per Sex", x=1, y=2.2, fontsize=20)
ax1 = fig.add_axes([0, 1, 1, 1], aspect=1)
ax1.title.set_text("Male")
ax2 = fig.add_axes([1, 1, 1, 1], aspect=1)
ax2.title.set_text("Female")
ax1.pie([m, 1-m], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
ax2.pie([f, 1-f], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
plt.show()
```

- 남성: 18.89%
- 여성: 74.20%

여성의 생존율이 매우 높고, 남성의 생존율이 매우 낮다.

## 4. 각 티켓 클래스별 성별에 따른 생존율 (원형 그래프)

```python
p1 = titanic_df[titanic_df.Pclass==1].groupby('Sex').Survived.mean()
p2 = titanic_df[titanic_df.Pclass==2].groupby('Sex').Survived.mean()
p3 = titanic_df[titanic_df.Pclass==3].groupby('Sex').Survived.mean()

fig = plt.figure()
fig.suptitle("Survival rate per Class & Sex", x=1, y=2.2, fontsize=20)

ax1 = fig.add_axes([0, 1, 1, 1], aspect=1)
ax1.title.set_text("C1 Male")
ax2 = fig.add_axes([1, 1, 1, 1], aspect=1)
ax2.title.set_text("C1 Female")
ax1.pie([p1.male, 1-p1.male], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
ax2.pie([p1.female, 1-p1.female], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')

ax3 = fig.add_axes([0, -.2, 1, 1], aspect=1)
ax3.title.set_text("C2 Male")
ax4 = fig.add_axes([1, -.2, 1, 1], aspect=1)
ax4.title.set_text("C2 Female")
ax3.pie([p2.male, 1-p2.male], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
ax4.pie([p2.female, 1-p2.female], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')

ax5 = fig.add_axes([0, -1.4, 1, 1], aspect=1)
ax5.title.set_text("C3 Male")
ax6 = fig.add_axes([1, -1.4, 1, 1], aspect=1)
ax6.title.set_text("C3 Female")
ax5.pie([p3.male, 1-p3.male], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')
ax6.pie([p3.female, 1-p3.female], labels=["Survived","Dead"], colors=["skyblue","red"], autopct='%1.2f%%')

plt.show()
```

- 남성의 생존율은 C1이 높고 C2와 C3가 비슷한 양상을 보인 반면,
- 여성의 생존율은 C1과 C2가 비슷한 양상으로 높고 C3가 떨어진다.

C1, C2의 생존율이 90% 이상으로 높다.
