```python
import matplotlib.pyplot as plt
```

## II. Matplotlib Case Study

### 일차함수 그래프

```python
plt.plot([1,2,3,4,5])
plt.show()
```

```python
plt.figure(figsize=(10,10)) #plot 전에 설정해야 한다.
plt.plot([1,2,3,4,5])

plt.show()
```

### 이차함수 그래프

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = x**2

y
```

```python
plt.plot(x, y)
plt.show()
```

#### 더 이쁘게, 곡선으로 그리자

```python
x = np.arange(-10,10,0.01)
y = x ** 2

x
```

```python
plt.plot(x, y)
plt.show()
```

#### 꾸미기

1. x, y축 라벨
1. both 축 범위
1. 눈금 (ticks)
1. 그래프 타이틀(title)

```python
#plt.plot(x, x**2)

plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5,5,0,30])

plt.xticks([i for i in range(-5, 6)])

plt.title("y=x^2 graph")

#plt.plot(x, x**2)   두 위치다 유효.
plt.show()
```

#### 꾸미기 2

1. plot에 라벨링
1. legend

```python
plt.plot(x, x**2, label="trend")
plt.show()
```

```python
plt.plot(x, x**2, label="trend")
plt.legend()
```

## III. Matplotlib Case Study

꺽은 선 그래프

```python
x = np.arange(20)
y = np.random.randint(0,101,20)

y
```

```python
plt.plot(x,y)
plt.yticks(np.arange(0,101,10))

plt.show()
```

```python
plt.axis([0, 20, 0, 90])
plt.xticks(np.arange(0,21))
plt.plot(x, y)

plt.show()
```

산점도

```python
plt.scatter(x,y)

plt.show()
```

박스 그림

```python
plt.boxplot((x,y)) # params는 tuple로
plt.show()
```

막대그래프

```python
plt.bar(x,y)
plt.show()
```

히스토그램

```python
plt.hist(y, bins=np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))

plt.show()
```

원형 그래프

```python
z = [10,20,30,40]

plt.pie(z)
plt.show()
```

```python
plt.pie(z, labels=["sea", "orange", "leaf", "apple"])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

```python
import seaborn as sns

x = np.arange(0,22,2)
y = np.random.randint(0,20,20)

print(y)

plt.hist(y, bins=x)
plt.show()
```

커널밀도그림

```python
sns.kdeplot(y, shade=True)
plt.show()
```

카운트그림

```python
import pandas as pd

d = {"name":["Andy","Bob","Cat"], "vote":[True, True, False]}
vote_df = pd.DataFrame(d)

vote_df
```

```python
vote_count = vote_df.groupby("vote").count()

vote_count
```

```python
plt.bar(x=[False,True], height=vote_count["name"])
plt.show()
```

```python
sns.countplot(x=vote_df['vote'])
plt.show()
```

캣그림

```python
covid = pd.read_csv('./country_wise_latest.csv')
covid.head(5)
```

```python
sns.catplot(x="WHO Region", y='Confirmed', data=covid)
plt.show()
```

```python
sns.stripplot(x="WHO Region", y="Confirmed", data=covid)
plt.show()
```

```python
sns.swarmplot(x="WHO Region", y="Confirmed", data=covid)
plt.show()
```

히트맵

```python
a = covid.corr()

a
```

```python
sns.heatmap(a)
plt.show()
```
