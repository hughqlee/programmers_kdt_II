```python
books = {
    "Available": [True, True, False],
    "Location": [102, 215, 323],
    "Genre": ["Programing", "Physics", "Math"]
}

books_df = pd.DataFrame(books)

books_df
```

```python
books_df = pd.DataFrame(books, index=['버그란 무엇인가', '두근두근 물리학', '미분해줘 홈즈'])

books_df
```

```python
books_df.loc["버그란 무엇인가"]
```

```python
a = books_df.loc["미분해줘 홈즈", "Available"]
b = books_df.loc["미분해줘 홈즈"].Available
c = books_df.loc["미분해줘 홈즈"]["Available"]

print(a, b, c)
```

```python
books_df.iloc[1, 0:2]
```

### groupby

```python
covid_by_region = covid["Confirmed"].groupby(by=covid["WHO Region"])

print(covid_by_region)
```

```python
covid_by_region.sum()
```

```python
covid_by_region.mean()
```

## Mission:
### 1. covid 데이터에서 100 case 대비 사망률(`Deaths / 100 Cases`)이 가장 높은 국가는?

```python
covid[covid["Deaths / 100 Cases"]==covid["Deaths / 100 Cases"].max()]["Country/Region"]
```

### 2. covid 데이터에서 신규 확진자가 없는 나라 중 WHO Region이 'Europe'를 모두 출력하면?  
Hint : 한 줄에 동시에 두가지 조건을 Apply하는 경우 Warning이 발생할 수 있습니다.

```python
zero_case = covid[covid["New cases"]==0]
zero_case[zero_case['WHO Region']=='Europe']
```

### 3. 다음 [데이터](https://www.kaggle.com/neuromusic/avocado-prices)를 이용해 각 Region별로 아보카도가 가장 비싼 평균가격(AveragePrice)을 출력하면?

```python
avocado = pd.read_csv('./avocado.csv')

ave_prices_by_region = avocado['AveragePrice'].groupby(by=avocado['region'])
ave_prices_by_region.max()
```
